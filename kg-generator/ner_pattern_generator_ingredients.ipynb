```python
# This module takes json inputs and outputs ner patterns 
import json

```

```python
# reading ingredients
with open('./data/ingredients.json') as json_file:
    ingredients = json.load(json_file)

# print(data)
```

```python
# generating formatted ingredients
formatted_ingredients = []
for ing in ingredients.values():
    # print(prop['name'])
    ing = ing['name']
    ing = ing.strip()
    ing = ing.lower()
    print(ing)
    if ing not in formatted_ingredients:
        formatted_ingredients.append(ing)

print(formatted_ingredients)
```

```python
patterns = []

for ing in formatted_ingredients:
    base = {"label": "AMING", "pattern": ""}
    base["pattern"] = ing
    # print(base)
    patterns.append(base)
# patterns
patterns = [json.dumps(pattern) for pattern in patterns]
print(patterns)

with open("./data/ingredients.jsonl", "w") as f: 
    # for pattern in patterns:
    for pattern in patterns:
        f.write(pattern)
        f.write('\n')
```
