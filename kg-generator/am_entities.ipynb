```python
import spacy
import scispacy
import pandas as pd

from spacy.pipeline import EntityRuler
# from config import OUTPUTS_FOLDER
OUTPUTS_FOLDER = "./data/"


```

```python
df = pd.read_csv(OUTPUTS_FOLDER + "inputs.csv")
df.head()
```

```python
# nlp = spacy.load("en_core_sci_lg")
nlp = spacy.load("en_core_web_sm")
properties_ruler = EntityRuler(nlp).from_disk("./data/am.jsonl")
nlp.add_pipe(properties_ruler)
```

```python
skipped = 0
labels = ['AMPROP', 'AMING', 'ENTITY', 'DATE', 'TIME', 'PERSON', 'CARDINAL',
        'QUANTITY', 'MONEY', 'PERCENT', 'WORK_OF_ART', 'GPE', 'NORP', 'ORG',
        'PRODUCT', 'EVENT', 'ORDINAL', 'FAC', 'LOC', 'LANGUAGE', 'LAW']
label_trackers = []
for index, row in df.iterrows():
    label_tracker = {}
    for label in labels:
        label_tracker[label] = []
    try:
        doc = nlp(row['raw_text'].lower())
        # print(label_tracker)
        for ent in doc.ents:
            label_tracker[ent.label_].append(ent.text)
    except:
        skipped = skipped + 1
    label_trackers.append(label_tracker)
# skipped
# label_trackers
```

```python
label_trackers_df = pd.DataFrame(label_trackers)
# label_trackers_df
```

```python
foo = pd.concat([df, label_trackers_df], axis=1)
foo.to_csv("data/entites.csv", index=False)
```
